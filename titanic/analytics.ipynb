```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
preprocessed_train = pd.read_csv('preprocessed_train.csv')
preprocessed_train.head(10)
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(
    preprocessed_train.astype(float).corr(),
    linewidths=0.1,
    vmax=1.0, 
    square=True,
    cmap=colormap,
    linecolor='white',
    annot=True
)
```

```python
g = sns.pairplot(
    preprocessed_train,
    hue='Survived',
    palette = 'seismic',
    size=1.2,
    diag_kind = 'kde',
    diag_kws=dict(shade=True),
    plot_kws=dict(s=10)
)
g.set(xticklabels=[])
```
